```python
#tf meminta label dalam bentuk one hot encoding
from tensorflow.keras.utils import to_categorical
cat_ytrain = to_categorical(ytrain)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model

tf.random.set_seed(1103)
```

```python
class Net(Model):
  def __init__(self):
    super(Net, self).__init__()
    self.d1 = Dense(3, activation = 'relu')
    self.d2 = Dense(3, activation = 'softmax')

  def call(self, x):
     x = self.d1(x)
     return self.d2(x)

modelkita = Net()
```

```python
#menggunakan algoritma SGD (Stochastic Gradient Descent)
optimizers = tf.keras.optimizers.SGD(learning_rate = 0.01)
```

```python
#gunain algoritma untuk menghitung error rate - crossentropy
criterion = tf.keras.losses.CategoricalCrossentropy()
```

Alur Training
1. Input data ke model untuk mendapatkan output model
2. hitung eror antara output dengan label
3. hitung gradient error
4. update bobot menggunakan algoritma optimasi (SGD)

50 epoch

```python
for epoch in range(50):
  with tf.GradientTape() as tape:
    out = modelkita(xtrain)
    loss = criterion(cat_ytrain, out)

    print("%d / 50 -- loss = %.4f " % (epoch, loss.numpy()))

  gradient = tape.gradient(loss, modelkita.trainable_variables)
  optimizers.apply_gradients(zip(gradient, modelkita.trainable_variables))
```

```python
#ambil nilai terbesar setiap kelasnya sebagai prediksi
prediction = tf.argmax(out, axis = 1)


#hitung hasil match dari hasil prediksi dengan label actual
total_benar = tf.reduce_sum(tf.cast(tf.equal(prediction, ytrain), tf.int32))

accuration = total_benar/len(cat_ytrain) *100
```

```python
print(accuration)
```
